```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)


```

```python
#Read in the data 
df=pd.read_csv(r'C:\Users\TEJA\Downloads\movies.csv\movies.csv')
```

```python
#Look at the data
df.head()
```

```python
#Data cleaning
#Lookin for missing values
df.isnull()
```

```python
#calculating the sum of null values
df.isnull().sum()
```

```python
#Filling with 0 in place of null values
df=df.fillna(value=0)
```

```python
df.isnull().sum()
```

```python
#Data types for columns
df.dtypes
```

```python
#Changing data type
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df

```

```python
df.dtypes
```

```python
#Sorting based on gross
df=df.sort_values(by=['gross'],ascending=False,inplace=False)
```

```python
df.head(5)
```

```python
#Drop duplicates
df.drop_duplicates()
```

```python
#Using scatter plot
plt.scatter(x=df['budget'],y=df['gross'])
plt.xlabel('Budget of the film')
plt.ylabel('Gross Earnings of the film')
plt.title('Budget vs Gross Earnings')
sns.set_style(style='whitegrid')
plt.show()
```

```python
#Using regression plot
sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"red"},line_kws={"color":"blue"},fit_reg=True)
```

```python
#lets start correlation
```

```python
# Select the numeric columns of interest
numeric_cols = ['budget', 'gross', 'runtime', 'score', 'votes', 'year']
numeric_df = df[numeric_cols]

```

```python

# Convert the columns to numeric type
numeric_df = numeric_df.apply(pd.to_numeric, errors='coerce')

```

```python

# Calculate the correlation matrix
numeric_df.corr(method='pearson')  #pearson,kendall,spearman
```

```python
#Using heat map
```

```python
corr_matrix = numeric_df.corr(method='pearson') 
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Feautures')
plt.xlabel('Movie Feautures')
plt.ylabel('Movie Feautures')
```

```python
correlation_mat=numeric_df.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.4]
high_corr
```
